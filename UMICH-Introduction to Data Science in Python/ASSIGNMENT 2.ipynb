```python
import pandas as pd


df = pd.read_csv("C:\\Users\\syjy0\\OneDrive\\바탕 화면\\전북대_산공\\GITHUB\Open-Course\\UMICH-Introduction to Data Science in Python\\olympics.csv", index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names = df.index.str.split('\s\(')

df.index = names.str[0] # the [0] element is the country name (new index) 
df['ID'] = names.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)
df = df.drop('Totals')


df
```

```python
# Q3. Which country has the biggest difference between their summer gold medal counts and winter gold medal counts relative to their total gold medal count?
#    Only include countries that have won at least 1 gold in both summer and winter.
gold=df[(df['Gold.1']>=1) & (df['Gold']>=1)]
gold['D']=abs((df['Gold']-df['Gold.1']))/df['Gold.2']

gold['D'].idxmax()
```

```python
# Q6. Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? 
# Use CENSUS2010POP.
# ANSWER: a list of string values.
```

```python
# sort_values: (정렬할 기준변수(왼쪽부터 차례대로 정렬)/ascednding= true:오름차순, false:내림차순)
q6=census.sort_values(['STNAME','CENSUS2010POP'],ascending=False).groupby('STNAME').head()
q6
```

```python
#위의 표에서groupby로 'STNAME'으로 정렬하고 sum으로 groupby에서 정리한 종류에 관해 다 더해서 정렬한다.=>위,아래 비교하며 이해할 것
q6=q6.groupby('STNAME').sum()
q6
```

```python
# 거의 마지막단계인 위 표에서 다시' CENSUS2010POP'에관해서 내림차순으로 정리/ head(3):위에서 3번째까지 정렬
q6=q6.sort_values(['CENSUS2010POP'],ascending=False).head(3)
q6
```

```python
#Q7. Which county has had the largest absolute change in population within the period 2010-2015? 
#    (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)
#    e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.
# ANSWER: a single string value.
```

```python
#정렬할 대상 축을 결정할 때에는 axis 를 이용합니다. 
#axis=0 라고 써주면 인덱스를 기준으로 정렬하며(기본값), 
#axis=1 라고 써주면 컬럼을 기준으로 정렬합니다.
#std: 표준편차
q7['STDEV'] = q7[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012',
                      'POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']].std(axis=1)
q7.loc[q7['STDEV'].idxmax()]['CTYNAME']
```
