```python
# Q1.-1
# Load the energy data from the file `Energy Indicators.xls`
# Call this DataFrame **energy**. 
# - The first two columns are unneccessary, so you should get rid of them.
# - change the column labels so that the columns are ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
# - rename 
#"Republic of Korea": "South Korea",
#"United States of America": "United States",
#"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
#"China, Hong Kong Special Administrative Region": "Hong Kong"

# Q1.-2
# Load the GDP data from the file `world_bank.csv`
# Call this DataFrame **GDP**. 
# Make sure to skip the header, and rename the following list of countries:
# "Korea, Rep.": "South Korea", 
# "Iran, Islamic Rep.": "Iran",
# "Hong Kong SAR, China": "Hong Kong"

# Q1-3
# Load the 'scimagojr-3.xlsx'
# Call this DataFrame **ScimEn**. 


#Join the three datasets: GDP, Energy, and ScimEn into a new dataset (using the intersection of country names). 
#Use only the last 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15). 
#The index of this DataFrame should be the name of the country, and the columns should be ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
#       'Citations per document', 'H index', 'Energy Supply',
#       'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008',
#       '2009', '2010', '2011', '2012', '2013', '2014', '2015'].


#ANSWER: DataFrame with 20 columns and 15 entries.
```

```python
import pandas as pd
import numpy as np
#엑셀파일을 보고 필요한 부분을 가져오도록 하자: 도구skip_footer/skiprows/parse_cols
#skipfooter: 파일의 맨 아래에서부터의 건너 뛸 줄 수
energy=pd.read_excel('C:\\Users\\syjy0\\OneDrive\\바탕 화면\\전북대_산공\\GITHUB\\Open-Course\\UMICH-Introduction to Data Science in Python/Energy Indicators.xls',skiprows=17,skipfooter=38)
energy=energy.drop(['Unnamed: 0','Unnamed: 1'],axis=1)
energy
```

```python
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy
```

```python
#F or all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.
energy.loc[energy['Energy Supply'] == '...'] = np.NaN

# .str.replace('찾을 내용`, `바꿀 내용`)+정규식 이용
#국가명 뒤에 (~~)붙은거 제거
energy['Country'] = energy['Country'].str.replace(r" \(.*\)","")

#국가명 뒤에 숫자 제거
energy['Country'] = energy['Country'].str.replace(r"([0-9]+)$","")

# .apply(pd.to_numeric): 문자열 칼럼을 숫자형으로 변환
energy[['Energy Supply', 'Energy Supply per Capita']] = energy[['Energy Supply', 'Energy Supply per Capita']].apply(pd.to_numeric)

# Convert `Energy Supply` to gigajoules,there are 1,000,000 gigajoules in a petajoule
energy['Energy Supply'] = energy['Energy Supply']*10**6

replace_dict={"Republic of Korea": "South Korea",
              "United States of America": 'United States',
              "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
              "China, Hong Kong Special Administrative Region": "Hong Kong"}
energy['Country'].replace(to_replace=replace_dict, inplace=True)

#The index of this DataFrame should be the name of the country
energy = energy.set_index('Country')

energy
```

```python
GDP=pd.read_csv('C:\\Users\\syjy0\\OneDrive\\바탕 화면\\전북대_산공\\GITHUB\\Open-Course\\UMICH-Introduction to Data Science in Python/world_bank.csv',skiprows=4)
replace_dict = {"Korea, Rep.": "South Korea", 
                "Iran, Islamic Rep.": "Iran",
                "Hong Kong SAR, China": "Hong Kong"}
GDP['Country Name'].replace(to_replace=replace_dict, inplace=True)
GDP=GDP.rename(columns={'Country Name':'Country'})

#The index of this DataFrame should be the name of the country
GDP = GDP.set_index('Country')

GDP
```

```python
ScimEn=pd.read_excel('C:\\Users\\syjy0\\OneDrive\\바탕 화면\\전북대_산공\\GITHUB\\Open-Course\\UMICH-Introduction to Data Science in Python/scimagojr-3.xlsx')

#The index of this DataFrame should be the name of the country
ScimEn = ScimEn.set_index('Country')

ScimEn
```

```python
#left_index=True, right_index=True:우리는 여기서 index를 3개의 DF다 'COUNTRY'로 지정
#따라서, '왼쪽(첫번째 DF),오른쪽(두번째 DF)의 색인으로 합치겠다'라는 뜻
Merge=pd.merge(energy,GDP,how='outer',left_index=True, right_index=True)
Merge2=pd.merge(Merge,ScimEn,how='outer',left_index=True, right_index=True)
Merge2=Merge2[['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
       'Citations per document', 'H index', 'Energy Supply',
       'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
Merge2
```

```python
#dropna함수는 column내에 NaN값이 있으면 해당 내용은 필요없다 간주하고 삭제해버린다.
#threshthreshold(임계값)을 설정할 수 있는데, NaN이 ~개 이상 있는 것에 대해서만 삭제해버린다는 의미
print(Merge2.shape)
print(Merge2.shape[1])
print(Merge2.shape[1]-10)

#2016 IranGDP가 Nan이나오기 때문에 2006~2015사이에 Nan이 나오는 것을 제외시켜줘야한다.
#shape(행,열) 따라서 shape(0)=행, shape(1)=열
Merge2=Merge2.dropna(thresh=Merge2.shape[1]-10)
print(Merge2.shape)
Merge2
```

```python
Merge2=Merge2.loc[Merge2['Rank']<=15]
Merge2
```

```python
Merge2=Merge2.sort_values('Rank',ascending=True)
Merge2
```

```python
#Q4 By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?
#Answer: a single number.
```

```python
lis=Merge2[['2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
re=Merge2['2015']-Merge2['2006']
result=Merge2[['2006','2015']]

#평균GDP와 변화량을 데이터프레임에 새롭게 추가
result['change']=re
result['avgGDP']=lis.mean(axis=1)
result=result.sort_values('avgGDP', ascending=False)
result
```
